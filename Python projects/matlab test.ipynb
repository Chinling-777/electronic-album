```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin
from math import pi
```

```python
xs=list(range(4))
ys=[4.5,6.0,1.2,5.4]
plt.plot(xs,ys)
plt.show
#Basic steps:
# 1. Add data.
# 2. Plot.
# 3. Show plot
'x and y should be of same number of elements'
```

```python
plt.plot(xs,ys,color="blue",linewidth=1.0,linestyle="-",label="solid")
#设置呈现区间
plt.xlim(1,2)
plt.ylim(8,7)
#设置间隔”stick“可理解为分割区间的小棍
plt.xticks(np.linspace(-4,4,9,endpoint=True))
plt.yticks(np.linspace(-1,1,5))
#xy坐标轴和表格标题
plt.xlabel( 't (s)')
plt.ylabel( 'value (NA)')  
plt.title( 'Three Curves' )
#添加图例
plt.legend(loc='upper left',frameon=True)
#保存位置、格式和分辨率
#plt.savefig(”filePath/ex2.png”,dpi=72)
```
