```python
#dict methods
dictionary={}
s="abcdefgh"
for i in range(len(s)):
    dictionary[i]=s[i]
print(dictionary)
```

```python
dictionary.clear()
print(dictionary)
```

```python
#GET methods
#.item() return a list of tuples which contains the keys and values pair by pair
print(dictionary.items())
print(dictionary.keys())
print(dictionary.values())
print(dictionary.get(0))
#they all return values, although dict is mutable
```

```python
#ADD and DELETE methods
dictionary[8]='i'#<=>.append() method in list
del dictionary[0]
print(dictionary)
```

```python
#traversal 遍历字典得到的是keys
for j in dictionary:
    print(j )
for j in dictionary.items():
    print(j[1])
```

```python
words=['red','green','yellow']
dict1={}
for word in words:
    dict1[word]=words.index(word)#list的index函数！！！！！！
print(dict1)
```

```python
#lambda:适用于简单一行数学公式计算，返回值即一行计算式的结果
function_name=lambda a:a+8
#                  input :ruturn and operation
print(function_name(1))
```

```python
numbers=['51','19','81']
numbers.sort()#无返回值，改变本身
print(numbers)
print(numbers.sort())
numbers.sort(key=lambda a:a[1])#sort函数的本质是一个一个元素输入再比较运算后输出，这里lambda的意义在于使输入仅为第二位，而非两个字符同时输入
print(numbers)#上面只能是key
```

```python
myD = { 'ong':4, 'liat':2, 'wee': 6}  #sorted函数不改变原值
myD2 = {}
for apple, pen in sorted(myD.items(), key = lambda x:x[1]):  #sort by value
    print ("%s:%s" % (apple, pen))  
    myD2[apple] = pen
print(myD)
```
